# 06 TableQA — 表の集計・計算（text-to-pandas）

RAG（検索）が苦手な **集計・計算・条件抽出** を、pandas コードを LLM に生成させて解く。
Excel / CSV / DataFrame を扱える（要 `pip install -e "..[office]"`）。

- 検索・ルックアップ → PagedRAG
- 合計/平均/件数/条件抽出などの計算 → TableQA

```python
import llmlab
llmlab.settings_form()   # チャットAPI があれば動く（埋め込み不要）
```

```python
import pandas as pd
df = pd.DataFrame({
    "支店": ["東京","東京","大阪","大阪"],
    "月":   [4,5,4,5],
    "売上": [100,120,80,90],
})
tq = llmlab.TableQA(df)          # ファイルなら llmlab.TableQA("売上.xlsx")
ans = tq.ask("東京支店の売上合計は？")
print(ans)                       # 回答 + 生成コード + 実行結果
```

```python
print(ans.result)                # 計算結果そのもの
print(tq.code("支店ごとの売上平均は？"))   # 生成コードだけ確認
```
