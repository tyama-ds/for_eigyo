# 05 MultiPaperRAG v1: 複数論文の横断比較

**広く探す → 論文ごとに深掘り → 突き合わせて比較** の3段で、複数論文を横断比較する。
表（数値）の論文間比較にも対応（図/グラフは v2 予定）。

- `compare(q)` … 候補論文を特定 → 論文ごとに深掘り → 統合
- `compare_table(metric)` … 各論文の表から該当数値を抽出して比較表（要 `pip install pdfplumber`）

```python
import llmlab

llmlab.settings_form()   # Base URL / API Key / Model / Embed（RAG用）を入力
```

```python
mp = llmlab.MultiPaperRAG()

# 論文（PDF）をまとめて取り込む。表は pdfplumber で抽出キャッシュされる
mp.add_papers("../papers")
mp.papers()
```

```python
# 探す→深掘り→比較（本文ベース）
print(mp.compare("提案手法の精度を論文間で比較して。データセットも明記して"))
```

```python
# 表の数値を論文横断で比較表に（要 pdfplumber）
print(mp.compare_table("ImageNet accuracy"))
```
