# 05 MultiPaperRAG v2: 複数論文の横断比較

**広く探す → 論文ごとに深掘り → 突き合わせて比較**。PDF/Word/Excel 対応、図理解(pics)、
深掘りエンジン・探索戦略の切替、`to_df()` 出力に対応。

任意依存（必要に応じて）: `pip install -e "..[tables,office,figures]"`、`pip install pandas`

```python
import llmlab
llmlab.settings_form()   # Base URL / API Key / Model / Embed（RAG用）を入力
```

```python
mp = llmlab.MultiPaperRAG(
    deep_engine="paged",        # "paged"(速い) / "book"(構造的に深掘り)
    locate_strategy="search",   # "search" / "summary" / "all"
    pics=False,                 # True で PDF の図を VLM が理解（要 pymupdf + 画像対応モデル）
    # vlm_model="your-vision-model",
)
mp.add_papers("../papers")     # PDF/Word/Excel をまとめて取り込み
mp.papers()
```

```python
cmp = mp.compare("提案手法の精度を論文間で比較して。データセットも明記して")
print(cmp)
```

```python
cmp.to_df()      # 論文ごとの結果を DataFrame に（要 pandas）
```

```python
# 表の数値を論文横断で比較表に（PDF=pdfplumber / Word=python-docx / Excel=openpyxl）
print(mp.compare_table("ImageNet accuracy"))
```
