# 03 PagedRAG / DocRAG: 文書をページ出典つきで問い合わせる（標準ベクトルRAG）

`docs/` に文書（PDF など）を置き、取り込んで質問する。回答には **書名・ページ番号** の出典が付く。
これは LlamaIndex ベースの素直なベクトル RAG。**論文 BookRAG とは別物**（論文忠実版は `04_bookrag.ipynb`）。

```python
import llmlab

llmlab.settings_form()  # Base URL / API Key / Model / Embed Model を入力
```

```python
rag = llmlab.PagedRAG()   # DocRAG は別名

rag.add_book("../docs/営業マニュアル.pdf", title="営業マニュアル")
# rag.add_books("../docs")          # フォルダをまとめて取り込む
rag.books()                          # 取り込み済み一覧
```

```python
print(rag.ask("返品の手順は？"))                    # 回答＋ページ出典
print(rag.ask("料金体系は？", title="営業マニュアル"))  # 特定文書に絞る
```
