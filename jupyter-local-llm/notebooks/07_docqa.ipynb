# 07 DocQA — 1文書を 散文=RAG / 表=TableQA に自動振り分け

同じファイルの中で、**本文（散文）は PagedRAG**、**表は抽出して TableQA** に回す。
`ask()` が質問を分類して自動ルーティング（`route="rag"/"table"` で強制も可）。

依存: `pip install -e "..[tables,office]"`（表抽出）＋ 本文RAG は埋め込みAPI が必要。

```python
import llmlab
llmlab.settings_form()   # Model（チャット）＋ Embed（本文RAG用）
```

```python
doc = llmlab.DocQA("../docs/report.pdf")   # PDF / Excel / Word / CSV
doc.table_names()                          # 抽出された表の一覧
```

```python
print(doc.ask("この資料の要点を3つ教えて"))   # → 本文RAG に自動ルーティング
```

```python
print(doc.ask("売上の合計と平均は？"))          # → 表TableQA に自動ルーティング
```

```python
# ルートを強制したいとき
print(doc.ask("四半期の推移", route="table").route)
print(doc.ask("背景を説明して", route="rag").route)
```
