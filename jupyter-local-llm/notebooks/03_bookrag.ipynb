# 03 BookRAG: 本・長尺PDF をページ出典つきで問い合わせる

`docs/` に本（PDF など）を置き、取り込んで質問する。回答には **書名・ページ番号** の出典が付く。
生成・埋め込みとも `configure()` の OpenAI 互換エンドポイントを使う。

```python
import llmlab

llmlab.settings_form()  # 未設定なら Base URL / API Key / Model / Embed Model を入力
```

```python
book = llmlab.BookRAG()

# 1冊ずつ取り込む（title 省略時はファイル名）
book.add_book("../docs/営業マニュアル.pdf", title="営業マニュアル")

# もしくはフォルダ内の本をまとめて取り込む
# book.add_books("../docs")

book.books()  # 取り込み済み一覧
```

```python
# 質問（回答＋ページ出典が表示される）
print(book.ask("返品の手順は？"))
```

```python
# 特定の本だけに絞って質問
print(book.ask("料金体系は？", title="営業マニュアル"))
```

```python
# 出典を構造化データとして扱う
ans = book.query("解約の条件は？", top_k=3)
print(ans.text)
for s in ans.sources:
    print(s.title, s.page, round(s.score or 0, 3))
```
